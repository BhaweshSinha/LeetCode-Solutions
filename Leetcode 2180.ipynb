## Code:

```python
def countEven(num):
    x = [int(_) for _ in str(num)]
    s = [i for i in range(1,num+1)]
    res = []
    for i in s:
        if int((sum([int(_) for _ in str(i)]))%2) == 0:
            res.append(i)
    return len(res)
```

```python
#Testcase-1
num1 = 4
print(countEven(num1))
```

# Solution Explanation:

### Approach

The problem is to count how many numbers from 1 to num have a sum of digits that is even. To solve this, the code iterates through all numbers from 1 to num, computes the sum of digits for each number, and checks if the sum is even. If it is, the number is added to a list. Finally, the length of this list, which represents the count of numbers with an even digit sum, is returned.

### Algorithm

Convert the input num into a list of its digits (not strictly necessary for the core logic).
Create a list s containing all numbers from 1 to num.
Initialize an empty list res to store numbers with even digit sums.
For each number in s:
- Convert the number to its digits and compute the sum of digits.
- If the sum modulo 2 is 0 (even), append the number to res.
Return the length of res.

### Time Complexity

Iterating through numbers from 1 to num takes O(n).
Summing digits of a number with d digits takes O(d), where d ≤ log₁₀(num).
Overall time complexity is O(n log n) in terms of the number of digits processed.

### Space Complexity
The list res stores all numbers with even digit sums, which in the worst case could be up to num/2.
Space complexity is O(n).
