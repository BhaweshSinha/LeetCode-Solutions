## Code:

```python
def judgeCircle(moves):
    return moves.count('U') == moves.count('D') and moves.count('L') == moves.count('R')
```

```python
#Testcase-1
moves1 = "UD"
print(judgeCircle(moves1))
```

```python
#Testcase-2
moves2 = "LL"
print(judgeCircle(moves2))
```

# Solution Explanation:

### Approach
The problem determines whether a robot that starts at the origin (0, 0) returns to the same position after executing a sequence of moves. Each move shifts the robot one unit in a specific direction:
- 'U' moves up
- 'D' moves down
- 'L' moves left
- 'R' moves right
To end at the origin, the number of upward moves must equal the number of downward moves, and the number of left moves must equal the number of right moves. The code counts how many times each direction appears and checks whether the vertical and horizontal movements cancel each other out.

### Algorithm
First, count how many times 'U' appears in the string.
Second, count how many times 'D' appears in the string.
Third, count how many times 'L' appears in the string.
Fourth, count how many times 'R' appears in the string.
Finally, return True if the number of 'U' moves equals 'D' moves and the number of 'L' moves equals 'R' moves; otherwise return False.

### Time Complexity
Each count() operation scans the string once, taking O(n) time.
Since four counts are performed, the total time complexity is O(n).

### Space Complexity
No extra data structures proportional to input size are used.
Space complexity is O(1).
