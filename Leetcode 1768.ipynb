## Code:

```python
def mergeAlternately(word1, word2):
    result = []
    i = j = 0
    while i < len(word1) or j < len(word2):
        if i < len(word1):
            result.append(word1[i])
            i += 1
        if j < len(word2):
            result.append(word2[j])
            j += 1
    main = "".join(result)
    return main
```

```python
#Testcase-1
word1 = "abc"
word2 = "pqr"
print(mergeAlternately(word1,word2))
```

```python
#Testcase-2
word1 = "ab"
word2 = "pqrs"
print(mergeAlternately(word1,word2))
```

```python
#Testcase-3
word1 = "abcd"
word2 = "pq"
print(mergeAlternately(word1,word2))
```

# Solution Explanation:

### Approach

The problem merges two strings by taking characters alternately from each string. Two pointers are used to keep track of the current position in each word. The algorithm continues until all characters from both strings are processed. At each step, if a character is available in word1, it is added to the result, followed by a character from word2 if available. This ensures the characters are merged in alternating order while also handling cases where one string is longer than the other.

### Algorithm

First, initialize an empty list to store the merged characters.
Second, set two pointers i and j to track positions in word1 and word2.
Third, iterate while at least one pointer is within the length of its string.
Fourth, if i is within bounds of word1, append the character to the result and increment i.
Fifth, if j is within bounds of word2, append the character to the result and increment j.
Finally, join the list into a string and return it.

### Time Complexity

The loop runs until all characters from both strings are processed.
Each character is visited exactly once.
Overall time complexity is O(n + m), where n is the length of word1 and m is the length of word2.

### Space Complexity

An additional list is used to store the merged characters.
The space used is proportional to the total length of both strings.
Space complexity is O(n + m).
