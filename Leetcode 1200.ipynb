## Code:

```python
def minimumAbsDifference(arr):
    arr.sort()
    diff = []
    for i in range(len(arr)-1):
        diff.append(arr[i+1]-arr[i])
    diff.sort()
    mdiff = diff[0]
    result = []
    for i in range(len(arr)-1):
        if int(arr[i+1]-arr[i]) == mdiff:
            result.append([arr[i],arr[i+1]])
    return result
```

```python
#Testcase-1
arr1 = [4,2,1,3]
print(minimumAbsDifference(arr1))
```

```python
#Testcase-2
arr2 = [1,3,6,10,15]
print(minimumAbsDifference(arr2))
```

```python
#Testcase-3
arr3 = [3,8,-10,23,19,-4,-14,27]
print(minimumAbsDifference(arr3))
```

# Solution Explanation:

### Approach

The task is to find all pairs of numbers in the array that have the smallest absolute difference. The first step is to sort the array so that numbers are arranged in increasing order. Once sorted, the smallest difference between any two values will always appear between neighboring elements rather than elements far apart.

After sorting, the difference between each pair of adjacent elements is calculated and stored. The smallest value among these differences is identified. Then, the array is scanned again to collect all neighboring pairs whose difference equals this minimum value. These pairs form the final result.

### Algorithm

First, sort the given array in ascending order.
Second, compute the differences between every pair of neighboring elements and store them.
Third, determine the smallest difference from the stored values.
Fourth, go through the sorted array again and select all adjacent pairs that have this smallest difference.
Finally, return the list of these pairs.

### Time Complexity

Sorting the array requires O(n log n) time.
Calculating differences and collecting the valid pairs each require O(n) time.
Overall time complexity is O(n log n).

### Space Complexity

An additional list is used to store the differences, which requires O(n) space.
The output list of pairs may also take up to O(n) space in the worst case.
So, the overall space complexity is O(n).
