## Code:

```python
def uniqueOccurrences(arr):
    a = list(set(arr))
    c = []
    for i in a:
        c.append(arr.count(i))
    return len(c) == len(set(c))
```

```python
#Testcase-1
arr1 = [1,2,2,1,1,3]
print(uniqueOccurrences(arr1))
```

```python
#Testcase-2
arr2 = [1,2]
print(uniqueOccurrences(arr2))
```

```python
#Testcase-3
arr3 = [-3,0,1,-3,1,1,1,-3,10,0]
print(uniqueOccurrences(arr3))
```

# Solution Explanation:

### Approach
The problem checks whether the number of occurrences of each element in the array is unique. First, the distinct elements of the array are extracted using a set. Then, for each unique element, its frequency in the original array is counted and stored in a list. If all frequencies are unique, the length of the frequency list will be equal to the length of the set of those frequencies. If they differ, it means at least two elements share the same occurrence count, and the function returns False.

### Algorithm
First, create a set from the array to get all unique elements.
Second, initialize an empty list to store occurrence counts.
Third, loop through each unique element and count how many times it appears in the original array.
Fourth, store each count in the list.
Finally, compare the length of the count list with the length of the set of counts and return whether they are equal.

### Time Complexity
Finding the count of each unique element takes O(n) time, and this is done for up to n unique elements in the worst case.
Overall time complexity is O(nÂ²).

### Space Complexity
A list is used to store the frequency of each unique element, which can take up to O(n) space in the worst case.
Space complexity is O(n).
