## Code:

```python
def replaceDigits(s):
    alpha_ele = [s[x] for x in range(0,len(s),2)]
    num_ele = [int(s[i]) for i in range(1,len(s),2)]
    alpha = [__ for __ in 'abcdefghijklmnopqrstuvwxyz']
    res = []
    for i in range(len(num_ele)):
        merge_alpha = alpha[alpha.index(alpha_ele[i])+num_ele[i]]
        alpha_ele[i] = alpha_ele[i]+merge_alpha
    return ''.join(alpha_ele)
```

# Solution Explanation:

### Approach

The function processes a string where characters at even indices are lowercase letters and characters at odd indices are digits. Each digit represents how far to shift the letter before it in the alphabet.
First, the function separates letters and digits into two different lists. Then, for every letter–digit pair, it finds the new shifted character by moving forward in the alphabet by the digit’s value. The original letter and its shifted version are combined and stored back. Finally, all transformed pairs are joined together into a single resulting string.

### Algorithm

First, extract all characters at even indices (letters) into a list.
Second, extract all characters at odd indices (digits) and convert them into integers.
Third, create a list of all lowercase alphabet letters for reference.
Fourth, iterate through each letter–digit pair:
- Find the position of the letter in the alphabet.
- Shift forward by the digit value to get a new letter.
- Concatenate the original letter and the shifted letter.
Finally, join all updated letter pairs into one string and return it.

### Time Complexity

Extracting letters and digits each takes O(n) time.
For each pair, searching the letter’s index in the alphabet takes O(26), which is constant.
Since this happens for roughly n/2 pairs, the total time complexity is O(n).

### Space Complexity

Lists are created to store letters, digits, and the alphabet.
Each of these uses space proportional to the input size (except the alphabet which is constant).
So the overall space complexity is O(n).
