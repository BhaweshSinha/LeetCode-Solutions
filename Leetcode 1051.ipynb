## Code:

```python
def heightChecker(heights):
    arrn = [x for x in heights]
    arrn.sort()
    c = 0
    for i in range(len(arrn)):
        if arrn[i] != heights[i]:
            c = c+1
        i = i+1
    return c
```

```python
#Testcase-1
heights_1 = [1,1,4,2,1,3]
print(heightChecker(heights_1))
```

```python
#Testcase-2
heights_2 = [5,1,2,3,4]
print(heightChecker(heights_2))
```

```python
#Testcase-3
heights_3 = [1,2,3,4,5]
print(heightChecker(heights_3))
```

# Solution Explanation:

### Approach
The problem checks how many elements in the given heights list are not in the position they would be if the list were sorted. To do this, a copy of the original list is created so the original order is preserved. The copied list is then sorted to represent the expected arrangement. Both lists are compared index by index, and each time the values at the same index differ, a counter is incremented. The final counter value represents the number of students standing in incorrect positions.

### Algorithm
First, make a copy of the input array.
Second, sort the copied array in non-decreasing order.
Third, iterate through both arrays simultaneously and count the indices where the values differ.
Finally, return the count.

### Time Complexity
Sorting the array takes O(n log n) time.
Comparing the elements takes O(n) time.
Overall time complexity is O(n log n).

### Space Complexity
An additional array of size n is used to store the sorted values.
Space complexity is O(n).
