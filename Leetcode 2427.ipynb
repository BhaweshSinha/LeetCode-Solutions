## Code:

```python
def commonFactors(a, b):
    lis = []
    c = 0
    if a>b:
        for i in range(1,b+1):
            lis.append(i)
    else:
        for j in range(1,a+1):
            lis.append(j)
    for k in lis:
        if a%k == 0 and b%k == 0:
            c = c+1
    return c
```

```python
# Testcase-1
print(commonFactors(12,6))
```

```python
# Testcase-2
print(commonFactors(25,30))
```

# Solution Explanation:

### Approach

The goal is to count how many positive integers divide both given numbers a and b.
To do this, the solution first determines the smaller of the two numbers, since a common factor cannot be greater than the smaller value. All integers from 1 up to this smaller number are considered as possible factors. Each candidate is checked to see whether it divides both numbers exactly. A counter is used to keep track of how many such common factors are found.

### Algorithm

First, compare a and b to determine the smaller value.
Second, create a list containing all integers from 1 to the smaller value (inclusive).
Third, iterate through this list and check whether each number divides both a and b without leaving a remainder.
Fourth, increment a counter whenever a common factor is found.
Finally, return the counter as the total number of common factors.

### Time Complexity

The loop runs up to min(a, b) times.
Each iteration involves constant-time modulo operations.
Therefore, the overall time complexity is O(min(a, b)).

### Space Complexity

An additional list of size min(a, b) is used to store candidate factors.
Hence, the space complexity is O(min(a, b)).
