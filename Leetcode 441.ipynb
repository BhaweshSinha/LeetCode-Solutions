## Code:

```python
def arrangeCoins(n):
    i = 0
    while n>=0:
        i+=1
        n = n-i
    return i-1
```

```python
#Testcase-1
n1 = 5
print(arrangeCoins(n1))
```

```python
#Testcase-2
n2 = 8
print(arrangeCoins(n2))
```

# Solution Explanation:

### Approach
The problem determines how many complete rows of a staircase can be built using n coins. The first row requires 1 coin, the second row requires 2 coins, the third row requires 3 coins, and so on. The code keeps subtracting the number of coins needed for each next row from the total n until there are not enough coins left to build another full row. The number of fully completed rows is then returned.

### Algorithm
First, initialize a counter i = 0 to represent the current row number.
Second, repeatedly increase the row number by 1.
Third, subtract the number of coins required for that row (i) from n.
Fourth, continue this process while n is still greater than or equal to 0.
Finally, when n becomes negative, return i - 1 because the last increment represents an incomplete row.

### Time Complexity
In the worst case, the loop runs until the sum of the first k natural numbers exceeds n.
Since 1+2+3+...+k=k(k+1)/2≈n1 + 2 + 3 + ... + k = k(k+1)/2 \approx n1+2+3+...+k=k(k+1)/2≈n, we get k≈2nk \approx \sqrt{2n}k≈2n​.
Therefore, the time complexity is O(√n).

### Space Complexity
The algorithm only uses a few integer variables (i and n) and does not use any extra data structures.
So, the space complexity is O(1).
