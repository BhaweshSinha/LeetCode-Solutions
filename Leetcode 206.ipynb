## Code:

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution(object):
    def reverseList(self, head):
        prev = None
        curr = head
        while curr is not None:
            next_node = curr.next
            curr.next = prev
            prev = curr
            curr = next_node
        return prev
```

```python
def build_linked_list(arr):
    head = ListNode(arr[0])
    curr = head
    for val in arr[1:]:
        curr.next = ListNode(val)
        curr = curr.next
    return head
```

```python
def print_linked_list(head):
    curr = head
    while curr:
        print(curr.val, end=" -> ")
        curr = curr.next
    print("None")
```

# Solution Explanation:

### Approach

The problem requires reversing a singly linked list so that the direction of all pointers is flipped. Instead of creating a new list, we reverse the links in-place. We use three pointers: one to track the previous node, one for the current node, and one to temporarily store the next node. As we traverse the list, we redirect each node’s next pointer to its previous node. By the end of the traversal, the previous pointer will be at the new head of the reversed list.

### Algorithm

First, initialize a pointer prev as None and set curr to the head of the list.
Second, iterate through the list while curr is not None.
Third, store the next node using a temporary pointer so the rest of the list is not lost.
Fourth, reverse the current node’s pointer by making it point to prev.
Fifth, move prev one step forward to curr, and move curr to the saved next node.
Finally, when the loop ends, return prev as the new head of the reversed list.

### Time Complexity

Each node in the linked list is visited exactly once.
Therefore, the time complexity is O(n), where n is the number of nodes.

### Space Complexity

No extra data structures are used; only a few pointer variables are maintained.
Therefore, the space complexity is O(1) (constant extra space).
